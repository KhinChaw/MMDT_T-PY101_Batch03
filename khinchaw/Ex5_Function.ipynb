Ex 3: Write a function called number_of_factors that takes an integer and returns how many
factors the number has.

```python
def number_of_factors(n):
    count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            count += 1
    return count

number_of_factors(5)
```

Ex4: Write a function called first_diff that is given two strings and returns the first location in
which the strings differ. If the strings are identical, it should return -1.Ignore the case in your function. It means h and H are the same.

```python
def first_diff(s1, s2):
    s1 = s1.lower()
    s2 = s2.lower()

    min_len = len(s1) if len(s1) < len(s2) else len(s2)

    for i in range(min_len):
        if s1[i] != s2[i]:
            return i

    if len(s1) != len(s2):
        return min_len
    return -1

print(first_diff("he", "she"))
print(first_diff("he", "her"))
print(first_diff("apple", "apply"))
print(first_diff("APPLE", "apple"))
print(first_diff("Hello", "hello"))
```

Ex6: Write a function that solves the system of equations ax + by = e and cx +dy = f. Your function should takes 6 parameters, (a, b, c, d, e,f) and return the values of x and y.

```python
def solve_system(a, b, c, d, e, f):
    # Compute the determinant
    det = a * d - b * c

    if det == 0:
        print("The system has no unique solution.")
        return None  # or raise an error

    # Use Cramer's rule to solve
    x = (e * d - b * f) / det
    y = (a * f - e * c) / det

    return x, y

# Example usage
x, y = solve_system(2, 3, 1, 2, 8, 5)
print("x =", x)
print("y =", y)
```

Ex7:  Write a function called change_case that given a string, returns a string with each upper  case letter replaced by a lower case letter and vice-versa

```python
def change_case(s):
    new_str = ""
    for char in s:
        if char.isupper():
            new_str += char.lower()
        elif char.islower():
            new_str += char.upper()
        else:
            new_str += char
    return new_str
```

```python
print(change_case("Python Programming"))
```

Ex8:  Write a function called primes that is given a number n and returns a list of the first n  primes. Let the default value of n be 100.

```python
def primes(n=100):

  if n <= 0:
    return []

  prime_list = []

  number = 2
  while len(prime_list) < n:

    is_prime = True

    for prime in prime_list:
      if number % prime == 0:
        is_prime = False
        break

    if is_prime:
      prime_list.append(number)

    number += 1

  return prime_list
```

```python
first_10_primes = primes(10)
print(first_10_primes)
```

Ex9: A Tic-tac-toe board can be represented be a 3×3 two-dimensional list, where zeroes stand for  empty cells, ones stand for X’s and twos stand for O’s. Write a function that is given such a list and randomly chooses a spot in which to place  a 2 (O). The spot chosen must currently be empty and a spot must be chosen.

```python
import random

def place_O(board):
    empty_positions = []
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                empty_positions.append((i, j))

    if not empty_positions:
        print("No empty spots available!")
        return

    row, col = random.choice(empty_positions)
    board[row][col] = 2
    print(f"Placed O at position ({row}, {col})")

board = [
    [1, 0, 2],
    [0, 1, 0],
    [0, 0, 1]
]

place_O(board)
for row in board:
    print(row)
```

Ex10:  Write a function that is given a 9 × 9 potentially solved Sudoku and returns True if it is solved correctly and False if there is a mistake. The Sudoku is correctly solved if there are
 no repeated numbers in any row or any column or in any of the nine “blocks.”


```python
def is_sudoku_solved(board):
    # Check rows
    for row in board:
        if sorted(row) != list(range(1, 10)):
            return False

    # Check columns
    for col in range(9):
        column = [board[row][col] for row in range(9)]
        if sorted(column) != list(range(1, 10)):
            return False

    # Check 3x3 blocks
    for block_row in range(0, 9, 3):
        for block_col in range(0, 9, 3):
            block = []
            for i in range(3):
                for j in range(3):
                    block.append(board[block_row + i][block_col + j])
            if sorted(block) != list(range(1, 10)):
                return False

    return True
```

```python
sudoku_board = [
    [5,3,4,6,7,8,9,1,2],
    [6,7,2,1,9,5,3,4,8],
    [1,9,8,3,4,2,5,6,7],
    [8,5,9,7,6,1,4,2,3],
    [4,2,6,8,5,3,7,9,1],
    [7,1,3,9,2,4,8,5,6],
    [9,6,1,5,3,7,2,8,4],
    [2,8,7,4,1,9,6,3,5],
    [3,4,5,2,8,6,1,7,9]
]

print(is_sudoku_solved(sudoku_board))  # Output: True
```
